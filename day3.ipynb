```python
#로또 사이트 불러오기
from urllib import request
from bs4 import BeautifulSoup
url_python = "https://dhlottery.co.kr/common.do?method=main"
target = request.urlopen(url_python)
soup = BeautifulSoup(target, "html.parser")
tag = soup.select("#numView")
tmp_list = tag[0].text.split("\n")
ball_list = []
for i in tmp_list:
    try: 
        ball_list.append(int(i))
    except:
        pass
ball_list

lottoN = soup.select("#lottoDrwNo")[0].text

print("{}회의 당첨번호 : {},{},{},{},{},{}\n보너스 번호는 {}".format(lottoN, ball_list[0],ball_list[1],ball_list[2],
                                                         ball_list[3],ball_list[4],ball_list[5],ball_list[6]))
```

```python
from urllib import request
from bs4 import BeautifulSoup
url_python = "https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
target = request.urlopen(url_python)
soup = BeautifulSoup(target, "html.parser")

for location in soup.select("location"):
    print("도시:", location.select_one("city").string)
    print("날씨:", location.select_one("wf").string)
    print("최저기온:",location.select_one("tmn").string)
    print("최고기온", location.select_one("tmx").string)
    print()
```
