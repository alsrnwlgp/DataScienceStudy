```python
import random
last_name = list("강김고이박최현한홍송봉백자")
first_name = ["민구", "민송", "정식", "윤회", "수정", "림수", "창훈", "재민", "창민", "현영", "준호", "준민", "민수", "철수", "영희"] 
with open("score_of_student.csv", "w") as file:
    sum_korean = 0
    sum_english = 0
    sum_math = 0
    for i in range(100):
        name = random.choice(last_name) + random.choice(first_name)
        korean = random.randrange(101)
        sum_korean += korean
        english = random.randrange(101)
        sum_english += english
        math = random.randrange(101)
        sum_math += math
        sum_score = korean + english + math
        average_score = sum_score / 3
        file.write("{}, {:3d}, {:3d}, {:3d}, {:3d}, {:3f}\n".format(name,korean,english,math,sum_score,average_score))
    file.write("  , {:3f}, {:3f}, {:3f}".format( sum_korean / 100, sum_english / 100, sum_math / 100))
with open("score_of_student.csv", "r") as file:
    print(file.read())
```
