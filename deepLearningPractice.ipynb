```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
print(diabetes.data.shape, diabetes.target.shape)
```

```python
plt.scatter(diabetes.data[:,3], diabetes.target)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

x_data = diabetes.data[:,3]
y_data = diabetes.target
```

```python
import numpy as np
#  
def distance_of_points_from_line(x_arr, y_arr, w, b):
    result = 0
    for i in range(len(x_arr)):
        result += (w*x_arr[i] - y_arr[i] + b)**2/w**2 + 1
    return result
    
def get_b_from_w(w, x_center, y_center):
    return y_center - w * x_center
    
def my_regression_function(x_arr, y_arr):
    x_center = sum(x_arr)/len(x_arr)
    y_center = sum(y_arr)/len(y_arr)
    w = 0
    b = get_b_from_w(w,x_center,y_center)
    distance = distance_of_points_from_line(x_arr, y_arr, w, b)
    where = "plus"
    tmp_w = w + 0.01
    tmp_b = get_b_from_w(tmp_w, x_center, y_center)
    tmp_distance = distance_of_points_from_line(x_arr, y_arr, tmp_w, tmp_b)
    if tmp_distance >= distance:
        where = "minus"
    #   
    print(where)
    while True:
        if where == "plus":
            tmp_w = w + 0.01
        else:
            tmp_w = w - 0.01
        tmp_b = get_b_from_w(tmp_w, x_center, y_center)
        tmp_distance = distance_of_points_from_line(x_arr, y_arr, tmp_w, tmp_b)
        if tmp_distance < distance:
            distance = tmp_distance
            w = tmp_w
            b = tmp_distance
        else:
            break
    return w, b
```
